```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

arquivo = pd.read_csv('b1abb33c4f.csv')
df = pd.DataFrame(arquivo)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize= (10,8))
fig, axes = plt.subplots(3,2)

plt.subplots_adjust(wspace=0.5, hspace=0.7)
sns.countplot(ax=axes[0][0], x='APGAR1', data=df)
sns.countplot(ax=axes[1][0], x='APGAR5', data=df)
sns.countplot(ax=axes[0][1], x='SEXO', data=df)
sns.countplot(ax=axes[1][1], x='GESTACAO', data=df)
sns.countplot(ax=axes[2][0], x='ESTCIVMAE', data=df)
axes[1][1].set_xticklabels(axes[1][1].get_xticklabels(), rotation=90)
fig.delaxes(axes[2][1])
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
df['porcentagem_apgar5']= df['APGAR5']/df['APGAR5'].sum()
df['porcentagem_apgar5'].value_counts()
```

```python
df['CONSULTAS'].value_counts()
```

```python
plt.rc('figure', figsize= (8,6))

sns.barplot( x='CONSULTAS',y='porcentagem_apgar5', data=df )
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize= (10,8))
fig, axes = plt.subplots(2,2)

plt.subplots_adjust(wspace=0.5, hspace=0.7)
sns.countplot(ax=axes[0][0], x='SEXO', data=df)
sns.countplot(ax=axes[1][0], x='GESTACAO', data=df)
axes[1][0].set_xticklabels(axes[1][0].get_xticklabels(), rotation=90)
sns.countplot(ax=axes[0][1], x='CONSULTAS', data=df)
sns.countplot(ax=axes[1][1], x='ESTCIVMAE', data=df)
```
